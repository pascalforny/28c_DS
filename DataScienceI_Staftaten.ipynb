```python
import pandas as pd
```

```python
#hier wird eine CSV-Datei geladen
dataframe = pd.read_csv("19-01-02_registrierte_Straftaten_einzeln.csv")
print(dataframe)
```

```python
#so können wir nur einen Teil der Daten auswerten
dataframe = dataframe.loc[dataframe['jahr'] == 2009]
```

```python
#So können wir den mean (Durchschnitt für alle numerischen Spalten berechnen)
dataframe.mean(numeric_only = True)
```

```python
#So können wir den Median berechnen
dataframe.median(numeric_only = True)
```
